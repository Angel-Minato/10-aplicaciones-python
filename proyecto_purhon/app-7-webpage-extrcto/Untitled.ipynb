```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

r=requests.get("https://pythonizing.github.io/data/real-estate/rock-springs-wy/LCWYROCKSPRINGS/")
c=r.content

soup = BeautifulSoup(c, "html.parser")

all=soup.find_all("div", {"class" : "propertyRow"})

all[0].find("h4").text.replace("\n","").replace(" ","")
```

```python
l=[]
base_url="https://pythonizing.github.io/data/real-estate/rock-springs-wy/LCWYROCKSPRINGS/t=0&s="
for pagina in range(0,30,10):
    print(base_url+str(pagina)+".html")
    c=r.content
    soup=BeautifulSoup(c,"html.parser")
    all=soup.find_all("div", {"class" : "propertyRow"})
    for i in all:
        d={}
        d["Address"]=i.find_all("span",{"class", "propAddressCollapse"})[0].text
        d["Locatizacion"]=i.find_all("span",{"class", "propAddressCollapse"})[1].text
        d["Price"]=i.find("h4", {"class" : "propPrice"}).text.replace("\n","").replace(" ","")
        try:
            d["Cama"]=i.find("span",{"class","infoBed"}).find("b").text
        except:
             d["cama"]=None
        try:
            d["Area"]=i.find("span",{"class","infoSqFt"}).find("b").text
        except:
            d["Area"]=None
        try:
            d["baño completo"]=i.find("span",{"class","infoValueFullBath"}).find("b").text
        except:
            d["baño completo"]=None
        try:
            d["medio baño"]=i.find("span",{"class","infoValueHalfBath"}).find("b").text
        except:
            d["medio baño"]=None
        for columns in i.find_all("div",{"class":"columnGroup"}):
            for featureGruop, featureName in zip(columns.find_all("span",{"class":"featureGroup"}),columns.find_all("span",{"class":"featureName"})):
               
                if "Lot Size" in featureGruop.text:
                    d["Gran tamaño"]=featureName.text
        
        l.append(d)
```
